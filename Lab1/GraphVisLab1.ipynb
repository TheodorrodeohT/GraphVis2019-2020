```
df_name_basics = pd.read_csv('./data/name.basics.tsv', sep='\t')
df_title_basics = pd.read_csv('./data/title.basics.tsv', sep='\t')
df_title_crew = pd.read_csv('./data/title.crew.tsv', sep='\t')
df_title_akas = pd.read_csv('./data/title.akas.tsv', sep='\t')
df_title_principals = pd.read_csv('./data/title.principals.tsv', sep='\t')
df_title_ratings = pd.read_csv('./data/title.ratings.tsv', sep='\t')
```

```
display(df_name_basics)
display(df_title_basics)
display(df_title_crew)
display(df_title_akas)
display(df_title_principals)
display(df_title_ratings)
```

```
df_title_crew_single_dir = df_title_crew[[len(x) == 9 for x in df_title_crew['directors']]]
df_result = df_title_crew_single_dir.join(df_title_basics.set_index('tconst'), on='tconst', how='inner').reset_index(drop=True)
df_result = df_result.join(df_title_ratings.set_index('tconst'), on='tconst', how='inner').reset_index(drop=True)
df_result = df_result.join(df_title_akas.set_index('titleId'), on='tconst', how='inner').reset_index(drop=True)
df_result = df_result.join(df_name_basics.set_index('nconst'), on='directors', how='inner').reset_index(drop=True)

df_result.drop(labels=[
                       'writers', 'endYear', 'birthYear', 'deathYear', 'primaryProfession', 'knownForTitles', 'primaryTitle', 'attributes', 'ordering', 'title'
                       ], axis=1, inplace=True)
df_result = df_result.drop_duplicates(subset=['originalTitle'], keep='first')

df_result = df_result[df_result['runtimeMinutes'] != '\\N']
df_result['runtimeMinutes'] = df_result['runtimeMinutes'].astype('int')

df_result.columns = ['titleId', 'directorId', 'titleType', 'title', 'isAdult', 'year', 'runtimeMins', 'genres', 'avgRating', 'numVotes', 'region', 'language', 'types', 'isOriginalTitle', 'director']
df_result = df_result[['titleId', 'title', 'directorId', 'director', 'year', 'titleType', 'isAdult', 'runtimeMins', 'genres', 'avgRating', 'numVotes', 'region', 'language', 'types', 'isOriginalTitle']]

df_result = df_result[(df_result['avgRating'] > 7.) & (df_result['numVotes'] > 100) & (df_result['runtimeMins'] > 60)]

df_result
```

```
from google.colab import files

df_result.to_csv('imdb_dataset_filtered.csv', index=False)
files.download('imdb_dataset_filtered.csv')
```

```
dirs = np.asarray(list(df_result.groupby('directorId').groups.keys()))

dirs_less = dirs[df_result.groupby('directorId')['titleId'].count() < 7]
dirs_more = dirs[df_result.groupby('directorId')['titleId'].count() > 2]

dirs_n = np.intersect1d(dirs_less, dirs_more)
dirs_n = np.random.choice(dirs_n, size=20)

df_result_visualise = df_result[df_result['directorId'].isin(dirs_n)]
df_result_visualise
```
